visualizing signals from multiple .mat files

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat

# Folder path containing all .mat files
base_dir = "/kaggle/input/raw-fetal-pcg-contaminated-with-mothers-pcg/Raw Fetal PCG Contaminated With Mother’s PCG"

def plot_mat_signals(file_path):
    data = loadmat(file_path)

    # Load variables (adjust if your file structure differs)
    Fs = float(data['Fs'])
    y = data['y']

    # Ensure correct orientation
    if y.shape[0] > y.shape[1]:
        y = y.T

    num_channels, num_samples = y.shape
    t = np.arange(num_samples) / Fs

    print(f"Plotting {os.path.basename(file_path)}: {num_channels} channels × {num_samples} samples (Fs={Fs} Hz)")

    fig, axes = plt.subplots(num_channels, 1, figsize=(12, num_channels * 2), sharex=True)
    if num_channels == 1:
        axes = [axes]

    for i, ax in enumerate(axes):
        ax.plot(t, y[i], color='black', linewidth=0.7)
        ax.set_ylabel(f"Ch {i+1}")
        ax.grid(True, linestyle=':', linewidth=0.5)
    
    axes[-1].set_xlabel("Time (s)")
    plt.suptitle(f"{os.path.basename(file_path)}")
    plt.tight_layout()
    plt.show()

# --- Loop through and plot all .mat files ---
mat_files = [os.path.join(base_dir, f) for f in os.listdir(base_dir) if f.endswith(".mat")]

print(f"Found {len(mat_files)} .mat files.\n")

for file_path in mat_files:
    plot_mat_signals(file_path)
```

Summary Statistics for Fetal PCG Dataset

```python
import os
import numpy as np
import pandas as pd
from scipy.io import loadmat

# Folder containing all your .mat files
base_dir = "/kaggle/input/raw-fetal-pcg-contaminated-with-mothers-pcg/Raw Fetal PCG Contaminated With Mother’s PCG"
# Collect summary stats
summary = []

for fname in os.listdir(base_dir):
    if fname.endswith(".mat"):
        path = os.path.join(base_dir, fname)
        data = loadmat(path)
        Fs = float(data['Fs'])
        y = data['y']
        if y.shape[0] > y.shape[1]:
            y = y.T
        
        summary.append({
            "subject": fname.replace(".mat", ""),
            "channels": y.shape[0],
            "samples": y.shape[1],
            "Fs": Fs,
            "duration_sec": y.shape[1] / Fs,
            "mean": np.mean(y),
            "std": np.std(y)
        })

df_summary = pd.DataFrame(summary)
print(df_summary.head())
```

```python
import matplotlib.pyplot as plt

# Mean comparison
plt.figure(figsize=(12,4))
plt.bar(df_summary['subject'], df_summary['mean'])
plt.xticks(rotation=90)
plt.ylabel("Mean Amplitude")
plt.title("Mean Signal per Subject")
plt.show()

# Std comparison
plt.figure(figsize=(12,4))
plt.bar(df_summary['subject'], df_summary['std'])
plt.xticks(rotation=90)
plt.ylabel("Std Deviation")
plt.title("Signal Variability per Subject")
plt.show()
```

outlier column flagging files detected by either IQR or MAD

```python
import numpy as np
import matplotlib.pyplot as plt

def find_outliers_iqr(df, column, multiplier=1.5):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - multiplier * IQR
    upper_bound = Q3 + multiplier * IQR
    outliers = (df[column] < lower_bound) | (df[column] > upper_bound)
    return outliers

def find_outliers_mad(df, column, threshold=3.5):
    values = df[column]
    median_val = np.median(values)
    mad = np.median(np.abs(values - median_val))
    modified_z_scores = 0.6745 * (values - median_val) / mad
    return np.abs(modified_z_scores) > threshold

# Columns to check for outliers
columns_to_check = ['mean', 'std', 'duration_sec', 'channels']

df_outliers = df_summary.copy()

for col in columns_to_check:
    df_outliers[f'{col}_outlier_iqr'] = find_outliers_iqr(df_outliers, col)
    df_outliers[f'{col}_outlier_mad'] = find_outliers_mad(df_outliers, col)

# Flag rows where any metric is outlier by either method
df_outliers['any_outlier'] = df_outliers[
    [f'{col}_outlier_iqr' for col in columns_to_check] +
    [f'{col}_outlier_mad' for col in columns_to_check]
].any(axis=1)

print("Summary of files flagged as outliers by any method:")
print(df_outliers[df_outliers['any_outlier']][
    ['subject'] +
    [f'{col}_outlier_iqr' for col in columns_to_check] +
    [f'{col}_outlier_mad' for col in columns_to_check]
])

# Visualization
for col in columns_to_check:
    plt.figure(figsize=(10, 4))
    plt.hist(df_summary[col], bins=30, alpha=0.7, label='All data')
    
    # IQR outliers
    iqr_out_vals = df_outliers.loc[df_outliers[f'{col}_outlier_iqr'], col]
    plt.scatter(iqr_out_vals, np.zeros_like(iqr_out_vals), color='red', label='IQR Outliers')
    
    # MAD outliers
    mad_out_vals = df_outliers.loc[df_outliers[f'{col}_outlier_mad'], col]
    plt.scatter(mad_out_vals, np.ones_like(mad_out_vals)*0.1, color='orange', label='MAD Outliers')
    
    plt.title(f'Distribution of {col} with IQR (red) and MAD (orange) Outliers')
    plt.xlabel(col)
    plt.legend()
    plt.show()
```

Euclidean Distance Between Subjects

```python
import os
import numpy as np
from scipy.io import loadmat
from sklearn.preprocessing import normalize
from scipy.spatial.distance import pdist, squareform
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the base directory where your .mat files are located
base_dir = "/kaggle/input/raw-fetal-pcg-contaminated-with-mothers-pcg/Raw Fetal PCG Contaminated With Mother’s PCG"
# List all .mat files in the directory
mat_files = [os.path.join(base_dir, f) for f in os.listdir(base_dir) if f.endswith('.mat')]
print(f"Found {len(mat_files)} .mat files.")

segments = []

for f in mat_files:
    data = loadmat(f)
    y = data['y']
    
    # Ensure rows = channels, columns = samples
    if y.shape[0] > y.shape[1]:
        y = y.T
    
    # Extract the first channel (all samples)
    first_channel = y[0, :]
    
    # Normalize to unit vector to focus on shape similarity
    first_channel_norm = normalize(first_channel.reshape(1, -1))[0]
    segments.append(first_channel_norm)

# Resample signals to the minimum length to have uniform size
min_len = min([s.shape[0] for s in segments])
segments_resampled = np.array([s[:min_len] for s in segments])

# Compute pairwise Euclidean distances
dist_matrix = squareform(pdist(segments_resampled, metric='euclidean'))

# Create a DataFrame with readable subject names
subject_names = [os.path.basename(f).replace(".mat", "") for f in mat_files]
df_dist = pd.DataFrame(dist_matrix, index=subject_names, columns=subject_names)

# Display the distance matrix
print(df_dist)

# Optional: Visualize the distance matrix with a heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df_dist, annot=True, fmt=".2f", cmap="viridis")
plt.title("Euclidean Distance Between Subjects (First Channel, All Samples)")
plt.show()
```

```python
import pandas as pd

df1 = pd.read_csv(
    '/kaggle/input/raw-fetal-pcg-contaminated-with-mothers-pcg/Raw Fetal PCG Contaminated With Mother’s PCG/Collected  data of batch 1 patients.csv',
    encoding='latin1'   # or 'ISO-8859-1'
)
```

Initial inspection

```python
print("Original shape:", df1.shape)
print("\nData Info:")
print(df1.info())

print("\nData Description:")
print(df1.describe(include='all'))

print("\nMissing values per column:")
print(df1.isnull().sum())

print("\nFirst 5 rows:")
print(df1.head())

# Optional: Plot histogram for numeric columns
import matplotlib.pyplot as plt
df1.select_dtypes(include=['number']).hist(figsize=(10,6))
plt.tight_layout()
plt.show()
```

Data cleaning

```python
# Step 1: Forward fill Patient_Weeks and Age_Weight since they relate to patient info across sessions
df1['Patient_Weeks'] = df1['Patient_Weeks'].ffill()
df1['Age_Weight'] = df1['Age_Weight'].ffill()

# Step 2: Extract numeric values from 'Mother_HeartRate' and 'Mother_Oxygen'
df1['Mother_HeartRate'] = df1['Mother_HeartRate'].str.extract('(\d+)').astype(float)
df1['Mother_Oxygen'] = df1['Mother_Oxygen'].str.extract('(\d+)').astype(float)

# Step 3: Parse 'Age_Weight' into separate numeric columns
df1[['Age_years', 'Weight_kg']] = df1['Age_Weight'].str.extract(r'(\d+)\s*years.*?(\d+)\s*kg').astype(float)

# Step 4: Clean 'Patient_Weeks' to get patient number and weeks
df1['Patient_Number'] = df1['Patient_Weeks'].str.extract(r'(\d+)').astype(int)
df1['Pregnancy_Weeks'] = df1['Patient_Weeks'].str.extract(r'\(([\d\+]+)').replace({'\+': ''}, regex=True).astype(float)

# Step 5: Check final cleaned data
print(df1.head())
print(df1.info())
```

visualizations

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df1['Mother_HeartRate'], kde=True)
plt.title("Mother's Heart Rate Distribution")
plt.show()
sns.histplot(df1['Fetus_HeartRate'], kde=True)
plt.title("Fetus Heart Rate Distribution - Batch 1")
plt.show()

sns.boxplot(x='Session_Per_Minute', y='Fetus_HeartRate', data=df1)
plt.title("Fetus Heart Rate by Session")
plt.show()
```

```python
corr = df1[['Mother_HeartRate', 'Mother_Oxygen', 'Age_years', 'Weight_kg', 'Fetus_HeartRate']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
